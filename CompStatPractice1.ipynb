```python
# computational statistics lab practice

#importing lib

import pandas as pd
import matplotlib.pyplot as pyplt

df = pd.read_excel('C:\\Users\\exam2\\Documents\\StudData.xlsx')
df
```

```python
df['total'] = df['Cs']+df['SE']+df['OS']
df['percentage'] = round(df['total']/3,2)
df
```

```python
def gradeCal(marks):
    if marks > 90: return 'A'
    elif marks > 80: return 'B'
    elif marks > 70: return 'C'
    elif marks > 60: return 'D'
    elif marks > 40: return 'E'
    else :return 'F'
```

```python
df['grade'] = [ gradeCal(i) for i in df['percentage']] 
df
```
